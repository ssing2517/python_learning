```python
import matplotlib.pyplot as plt
s1_marks=[20,40,50,30]
s2_marks=[23,38,40,20]
subject=['english','maths','science','hindi']
plt.scatter(s1_marks,subject,color='green',label='student_1 marks')
plt.scatter(s2_marks,subject,color='blue',label='student_2 marks')
#plt.show()-removes <matplotlib.collections.PathCollection at 0x793fb11335d0> from output
plt.xlabel('students marks')
plt.ylabel('subject')
plt.title('marks vs subject')
plt.legend()
plt.show()
```
